```python
#Drop unnecessary columns that do not contribute to the analysis.
keep_cols = [
    'Total Current Assets', 'Total current Liabilities', 'Cash & Bank',
    'Inventory', 'Accounts Receivable',
    'Total Assets', 'Total Liabilities', 'Total Equity',
    'Net Revenue', 'COSG', 'Interest expenses', 'Net profit after tax'
]

df_clean = df[keep_cols]
```

```python
df_clean = df_clean.rename(columns={
    'Total Current Assets': 'T_Current_Assets','Total current Liabilities':'T_Current_Liabilities','Cash & Bank':'Cash','Accounts Receivable':'Accounts_Receivable',
    'Total Assets':'T_Assets','Total Liabilities':'T_Liabilities','Total Equity':'T_Equity','Net Revenue':'Net_Revenue','Interest expenses':'IE',
    'Net profit after tax':'NPAT','COSG':'COGS'
})
```

```python
df_clean = df_clean.replace({' ': '', ',': ''}, regex=True)

df_clean = df_clean.replace('-', '0', regex=True)
```

```python
print(df_clean.dtypes)
```

```python
df_clean.info()
```

```python
col_name = [
    'T_Current_Assets','T_Current_Liabilities','Cash','Inventory',
    'Accounts_Receivable','T_Assets','T_Liabilities','T_Equity',
    'Net_Revenue','COGS','IE','NPAT'
]

for col in col_name:
    df_clean[col] = (
        df_clean[col]
        .astype(str)
        .str.replace('.', '', regex=False) # remove dots safely
        .str.replace('-','', regex=False)   
        .str.strip()
    )

    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

    
```

```python
# Prevent zero / near-zero denominators
df_clean['Inventory'] = df_clean['Inventory'].mask(df_clean['Inventory'] < 1, np.nan)
df_clean['COGS'] = df_clean['COGS'].mask(df_clean['COGS'] < 1, np.nan)
df_clean['T_Current_Liabilities'] = df_clean['T_Current_Liabilities'].mask(df_clean['T_Current_Liabilities'] < 1, np.nan)
df_clean['Accounts_Receivable'] = df_clean['Accounts_Receivable'].mask(df_clean['Accounts_Receivable'] < 1, np.nan)
df_clean['T_Equity'] = df_clean['T_Equity'].mask(df_clean['T_Equity'] < 1, np.nan)
```

```python
df_clean.info()
```

Interest expense in the raw dataset has too many missing values, even though it is a very important figure for calculating interest coverage. I had to drop this column.

```python
df_clean = df_clean.drop(columns=['IE'])
df_clean.info()
```

Because the dataset is messy and contains invalid values (less than or equal to 0), I applied filtering conditions to remove those rows.

```python
condition_garbage = (df_clean['T_Assets'] <= 0) | (df_clean['Net_Revenue'] <= 0)

df_garbage = df_clean[condition_garbage]

df_clean = df_clean[~condition_garbage].copy()
```

Another issue was records where total assets did not equal the sum of total liabilities and total equity, so I applied filtering conditions to remove those rows.

```python
# Set tolerance deviation to 1000 VND
tolerance = 1000 

# Check the deviation of value
df_clean['Balance_Diff'] = abs(df_clean['T_Assets'] - (df_clean['T_Liabilities'] + df_clean['T_Equity']))

# Remove invalid rows from the dataset
df_clean = df_clean[df_clean['Balance_Diff'] <= tolerance].copy()
```

```python
#Start with Efficiency Ratios#
df_clean['Inv_Turn'] = df_clean['COGS']/df_clean['Inventory'].replace(0, np.nan)
df_clean['DSI'] = 365 * df_clean['Inventory'] / df_clean['COGS'].replace(0, np.nan)
#Profitability Ratios"
df_clean['ROA'] = df_clean['NPAT']/df_clean['T_Assets'].replace(0, np.nan)
df_clean['ROE'] = df_clean['NPAT']/df_clean['T_Equity'].replace(0, np.nan)
df_clean['Cur_Ratio'] = df_clean['T_Current_Assets']/df_clean['T_Current_Liabilities'].replace(0, np.nan)
df_clean['Q_Ratio'] = (df_clean['T_Current_Assets']-df_clean['Inventory'])/df_clean['T_Current_Liabilities'].replace(0, np.nan)
df_clean['Cas_Ratio'] = df_clean['Cash']/ df_clean['T_Current_Liabilities'].replace(0, np.nan)
df_clean['Receivable_Turn'] = df_clean['Net_Revenue']/df_clean['Accounts_Receivable'].replace(0, np.nan)
#Leverage Ratios"
df_clean['DE'] = df_clean['T_Liabilities']/df_clean['T_Equity'].replace(0, np.nan)
df_clean['Debt_Ratio'] = df_clean['T_Liabilities']/df_clean['T_Assets'].replace(0, np.nan)
```

```python
df_clean = df_clean.replace([np.inf, -np.inf], np.nan)
```

```python
missing_count = df_clean.isna().sum()
print(missing_count)
```

Since the Inventory Turnover and Receivables Turnover, DSI, Current Ratios, Quick Ratios, Cash Ratios columns contain only a small number of missing values relative to the entire dataset, I will drop those rows.

```python
df_clean = df_clean.dropna()
```

The DSI formula is correctly applied; however, extremely high values arise when the cost of goods sold (COGS) is very small, leading to inflated ratios. To address this, observations with near-zero denominators are treated and extreme values are capped to ensure meaningful interpretation.

```python
df_clean['DSI'] = df_clean['DSI'].clip(upper=365)
```

As we can see, the distribution reflects a typical economic reality: the dense cluster on the left represents the majority of SMEs in Vietnam Sewing Industry, while the heavy right tail is driven by a small number of large corporations with assets or revenue significantly higher than the mean.

To mitigate the distortion caused by outliers, the Interquartile Range (IQR) method was applied with a sensitivity coefficient of $k = 3.0$. This approach effectively filters out extreme anomalies while preserving the data of high-performing, 'excellent' companies that represent the upper tier of the industry.

```python
def clean_financial_outliers_fixed(df_clean):
    df_out = df_clean.copy()
    # Create the first mask where every row is True
    is_not_outlier = pd.Series(True, index=df_out.index)
    
    for col in [('Cur_Ratio'), ('Q_Ratio'), ('Cas_Ratio'), 
                   ('Inv_Turn'), ('Receivable_Turn'), ('DSI'),
                   ('ROA'), ('ROE')]:
        
        if col in df_out.columns:
            Q1 = df_out[col].quantile(0.25)
            Q3 = df_out[col].quantile(0.75)
            IQR = Q3 - Q1
            
            # Updating mask: So if any row does not achive the additional the condition will be removed
            condition = (df_out[col] >= Q1 - 3.0*IQR) & (df_out[col] <= Q3 + 3.0*IQR)
            is_not_outlier &= condition
            
    return df_out[is_not_outlier]


df_final_clean = clean_financial_outliers_fixed(df_clean)
```

## In financial data analysis, "dirty data" is significantly more damaging than "limited data." 
* The Risk of the Full Dataset (1,500 rows): While providing a larger sample size, extreme anomalies (e.g., Cur_Ratio = 91,796) and contain accounting discrepancies where Total Assets does not equal Total Sum of Total Equity and Liabilities, would heavily distort sector averages. This would lead to skewed visualizations (Heatmaps) and biased statistical conclusions.

* The Benefit of the Refined Dataset (856 rows): These records represent "economic reality." By filtering for logical financial ranges, we ensure that correlation coefficients and trends are reliable and statistically significant.

I will categorize company size based on total assets, as the sewing industry relies heavily on physical resources such as machinery and inventory to operate. While total assets may not directly measure production capacity, they can serve as a reasonable proxy for a company’s operational scale in this industry.

```python
df_rank = df_final_clean.copy()

ranks = []
for index, row in df_rank.iterrows(): 
    assets = row['T_Assets']
    if assets > 500000: ranks.append('LARGE')
    elif 200000 < assets < 500000: ranks.append('BIG')
    else: ranks.append('SME')   

df_final_clean['Rank'] = ranks
```
